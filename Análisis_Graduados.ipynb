```python
import matplotlib.pyplot as plt
programa['INSTITUCIÓN DE EDUCACIÓN SUPERIOR (IES)'].value_counts().plot(kind='bar')

# Configurar etiquetas y título
plt.xlabel('Universidades')
plt.ylabel('Número de Matriculados')
plt.title('Distribución de estudiantes')

# Mostrar el gráfico
plt.show()
```

```python
df = pd.DataFrame(programa)
df
```

```python
# Graficar un scatter plot con Plotly Express
fig = px.bar(df['INSTITUCIÓN DE EDUCACIÓN SUPERIOR (IES)'].value_counts(),
             x=df['INSTITUCIÓN DE EDUCACIÓN SUPERIOR (IES)'].value_counts().index,
             y=df['INSTITUCIÓN DE EDUCACIÓN SUPERIOR (IES)'].value_counts().values,
             labels={'x': 'Universidades', 'y': 'Número de Matriculados'},
             title='Frecuencia de Universidades')

fig.update_xaxes(categoryorder='total descending', tickangle=20)
fig.show()
```

```python
fig = px.bar(df['PROGRAMA ACADÉMICO'].value_counts(),
             x=df['PROGRAMA ACADÉMICO'].value_counts().index,
             y=df['PROGRAMA ACADÉMICO'].value_counts().values,
             labels={'x': 'Programas', 'y': 'Número de Matriculados'},
             title='Frecuencia de Programas')

fig.update_xaxes(categoryorder='total descending', tickangle=20)
fig.show()
```
