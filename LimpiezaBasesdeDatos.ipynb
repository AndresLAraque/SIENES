```python
import pandas as pd
```

```python
archivo_excel_01 = '/content/drive/MyDrive/SNIES/Administrativos 2022.xlsx'
archivo_excel_02 = '/content/drive/MyDrive/SNIES/Docentes 2022.xlsx'
archivo_excel_03 = '/content/drive/MyDrive/SNIES/Estudiantes Admitidos 2022.xlsx'
archivo_excel_04 = '/content/drive/MyDrive/SNIES/Estudiantes Graduados 2022.xlsx'
archivo_excel_05 = '/content/drive/MyDrive/SNIES/Estudiantes Inscritos 2022.xlsx'
archivo_excel_06 = '/content/drive/MyDrive/SNIES/Estudiantes Matriculados 2022.xlsx'
archivo_excel_07 = '/content/drive/MyDrive/SNIES/Estudiantes Matriculados en primer curso 2022.xlsx'
archivo_excel_08 = '/content/drive/MyDrive/SNIES/Metadatos bases consolidadas 2023.xlsx'
```

```python
data = pd.read_excel(archivo_excel_01)
```

```python
# Elimina las primeras 4 filas
data.drop(range(4), axis=0, inplace=True)
```

```python
# Restablece el índice después de eliminar las filas
data = data.reset_index(drop=True)
```

```python
# Establece la primera fila como el encabezado
data.columns = data.iloc[0]
```

```python
# Elimina la primera fila después de establecerla como encabezado
data = data[1:].reset_index(drop=True)
```

```python
data
```

```python
# Verifica si hay valores NaN en el DataFrame
nan_values = data.isna().any().any()
```

```python
# Verifica si hay valores nulos en el DataFrame
#nan_values = data.isnull().any().any()
```

```python
if nan_values:
  print("EL dataframe tiene valores NaN o vacios")
else:
  print("El dataframe no tiene valores NaN o vacios.")
```

```python
# Exporta el DataFrame a un archivo CSV
data.to_csv('/content/drive/MyDrive/SNIES/Administrativos2022.csv', index=False)
```
