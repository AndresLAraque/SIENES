```python
abreviaturas = {'MAESTRÍA EN INGENIERÍA DE INFORMACIÓN': 'MINE.', 'MAESTRIA EN INTELIGENCIA ANALITICA DE DATOS': 'MIAD.', 'MAESTRÍA EN INTELIGENCIA ANALÍTICA PARA LA TOMA DE DECISIONES': 'M INT ANALÍTICA TD.', 'MAESTRÍA EN GESTIÓN DE INFORMACIÓN': 'M GEST INFO', 'MAESTRÍA EN CIENCIA DE DATOS': 'M CIENCIA DE DATOS.', 'MAESTRÍA EN INTELIGENCIA ARTIFICIAL': 'M INT ARTIFICAL', 'MAESTRIA EN INTELIGENCIA ARTIFICIAL': 'M INT ART.', 'MAESTRÍA EN INGENIERÍA Y ANALÍTICA DE DATOS': 'M ING Y ANA DATOS', 'MAESTRÍA EN CIENCIA DE DATOS': 'M CIENC DATOS', 'MAESTRÍA EN CIENCIAS DE LOS DATOS Y ANALÍTICA': 'M CIENCIA DAT Y ANA.'}
```

```python
abreviaturas_universidades = {'UNIVERSIDAD DE LOS ANDES': 'UNIANDES','UNIVERSIDAD ESCUELA COLOMBIANA DE INGENIERIA JULIO GARAVITO' : 'ESCUELA','PONTIFICIA UNIVERSIDAD JAVERIANA' : 'JAVERIANA','UNIVERSIDAD SERGIO ARBOLEDA': 'SERGIO','FUNDACION UNIVERSIDAD DE BOGOTA - JORGE TADEO LOZANO': 'U. TADEO','ICESI': 'ICESI','UNIVERSIDAD EAFIT-': 'ICESI'}
```

```python
df['PROGRAMA ACADÉMICO ABREVIADO'] = df['PROGRAMA ACADÉMICO'].map(abreviaturas)
```

```python
df['UNIVERSIDADES ABREVIADAS'] = df['INSTITUCIÓN DE EDUCACIÓN SUPERIOR (IES)'].map(abreviaturas_universidades)
```

```python
df_filtrado = df[df['PROGRAMA ACADÉMICO'].isin(programas_seleccionados)]
```

```python
df_agrupado = df_filtrado.groupby(['UNIVERSIDADES ABREVIADAS', 'PROGRAMA ACADÉMICO ABREVIADO'])['MATRICULADOS'].sum().reset_index()
```

```python
df_agrupado
```

```python
df_agrupado['PROGRAMA'] = df_agrupado['UNIVERSIDADES ABREVIADAS'] + '.' + df_agrupado['PROGRAMA ACADÉMICO ABREVIADO']
```

```python
df_agrupado
```

```python
plt.bar(df_agrupado['PROGRAMA'], df_agrupado['MATRICULADOS'])
plt.xlabel('Programa Académico')
plt.ylabel('Cantidad de Estudiantes Matriculados')
plt.title('Cantidad de Estudiantes Matriculados por Programa Académico')
plt.xticks(rotation='vertical', ha='center', va='top')
plt.show()
```

```python
import plotly.graph_objects as go

# Crear la figura de barras
fig = go.Figure()

# Agregar la barra al gráfico
fig.add_trace(go.Bar(
    x=df_agrupado['PROGRAMA'],
    y=df_agrupado['MATRICULADOS'],
    marker_color='blue'  # Puedes ajustar el color de las barras según tus preferencias
))

# Configurar el diseño del gráfico
fig.update_layout(
    xaxis=dict(
        title='Programa Académico',
        tickangle=-45,  # Rotar las etiquetas del eje x para mejorar la legibilidad
        tickmode='array',
        tickvals=list(range(len(df_agrupado['PROGRAMA']))),
        ticktext=df_agrupado['PROGRAMA'],
    ),
    yaxis=dict(title='Cantidad de Estudiantes Matriculados'),
    title='Cantidad de Estudiantes Matriculados por Programa Académico',
)

# Mostrar el gráfico
fig.show()
```
